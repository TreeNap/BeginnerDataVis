```python
from bokeh.io import show, output_notebook, curdoc
from bokeh.plotting import figure
from ipywidgets import interact
from bokeh.layouts import row, column, widgetbox, gridplot
from bokeh.models import ColumnDataSource
from bokeh.models.widgets import Slider, TextInput, Button, Paragraph, Select

output_notebook()
```

```python
#want graph year on x axis, y axis be value, legend be which subject of meat(or do graph for each meat) - interact by selecting country
import pandas as pd
df = pd.read_csv('meat_consumption.csv')
df.columns = ['location', 'indicator', 'subject', 'measure', 'frequency','time','value','flag_codes']

df_countries=df['location'].unique()
list_countries=list(df_countries)
df_measurements=df['measure'].unique()
list_measurements=list(df_measurements)
var="'USA'"
df_USA=df.query("location=={}".format(var))
df_USA_po=df_USA.query("subject=='POULTRY'and measure=='KG_CAP'")
df_USA_be=df_USA.query("subject=='BEEF'and measure=='KG_CAP'")
df_USA_sh=df_USA.query("subject=='SHEEP'and measure=='KG_CAP'")
df_USA_pi=df_USA.query("subject=='PIG'and measure=='KG_CAP'")

x=df_USA.time
USA_po=df_USA_po.value
USA_be=df_USA_be.value
USA_sh=df_USA_sh.value
USA_pi=df_USA_pi.value
```

```python

def modify_doc(doc):
    
    source_avg = ColumnDataSource(data=dict(
                                       x=x,
                                       y1=USA_po,
                                       y2=USA_be,
                                        y3=USA_sh,
                                        y4=USA_pi,
                                        y5=USA_po,
                                       y6=USA_be,
                                        y7=USA_sh,
                                        y8=USA_pi))
    
    TOOLS = 'tap'
    
    plot = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Poultry')
    
    plot.circle(x='x', y='y1', size=5, color='orange', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot.line(x='x', y='y1', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    
    plot2 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Beef')
    
    plot2.circle(x='x', y='y2', size=5, color='red', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot2.line(x='x', y='y2', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    plot3 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Sheep')
    
    plot3.circle(x='x', y='y3', size=5, color='gray', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot3.line(x='x', y='y3', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    plot4 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Pork')
    
    plot4.circle(x='x', y='y4', size=5, color='pink', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot4.line(x='x', y='y4', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    #2nd slot for comparison
    plot5 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Poultry')
    
    plot5.circle(x='x', y='y5', size=5, color='orange', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot5.line(x='x', y='y5', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    
    plot6 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Beef')
    
    plot6.circle(x='x', y='y6', size=5, color='red', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot6.line(x='x', y='y6', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    plot7 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Sheep')
    
    plot7.circle(x='x', y='y7', size=5, color='gray', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot7.line(x='x', y='y7', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    plot8 = figure(plot_width=300, plot_height=200, tools=TOOLS, toolbar_location=None, title='Pork')
    
    plot8.circle(x='x', y='y8', size=5, color='pink', source=source_avg,
                  hover_fill_color="black",
                  fill_alpha=0.7, hover_alpha=1,
                  line_color=None, hover_line_color="white") #fill_color="blue"
    
    plot8.line(x='x', y='y8', source=source_avg, line_dash="4 4", line_width=1, color='gray')
    
    
    # widgets for input
    t = Paragraph()
    text = TextInput(title="Title", value='USA')
    stat_select = Select(title="Countries", value="USA", options=list_countries)
    point_select = Select(title="Units of measurement", value="KG_CAP", options=list_measurements)
    era_select = Select(title="Date of measurement", value="All", options=["All","Past","Projected"])
    # widgets for input comparison
    text2 = TextInput(title="Title", value='USA')
    stat_select2 = Select(title="Countries", value="USA", options=list_countries)
    point_select2 = Select(title="Units of measurement", value="KG_CAP", options=list_measurements)
    #era_select2 = Select(title="Date of measurement", value="All", options=["All","Past","Projected"])

    # Set up callbacks
    def update_title(attrname, old, new):
        t.text = text.value

    text.on_change('value', update_title)
    
    def update_data_source(attrname, old, new):

        # Current selected values
        m = stat_select.value
        a = point_select.value
        e = era_select.value
        
        s = stat_select2.value
        u = point_select2.value
        #b = era_select2.value
        #code for unit of measurement
        if a=='KG_CAP':
            a_exp=' kilograms of retail weight per capita'
        else:
            #one other option
            a_exp=' thousand tonnes of carcass weight'
        
        # convert select values to format for query
        t.text = "{} vs {} consumption of meat in {}".format(m,s,a_exp)
        m="'{}'".format(m)
        a="'{}'".format(a)
        
        #code for changing era
        if e=="Past":
            mdf=df.query("location=={} and value>0 and time<2020".format(m))
        elif e=="Projected":
            mdf=df.query("location=={} and value>0 and time>2020".format(m))
        else:
            mdf=df.query("location=={} and value>0".format(m))

        mdf_po=mdf.query("subject=='POULTRY'and measure=={}".format(a))
        mdf_be=mdf.query("subject=='BEEF'and measure=={}".format(a))
        mdf_sh=mdf.query("subject=='SHEEP'and measure=={}".format(a))
        mdf_pi=mdf.query("subject=='PIG'and measure=={}".format(a))

        mx=mdf.time
        po=mdf_po.value
        be=mdf_be.value
        sh=mdf_sh.value
        pi=mdf_pi.value
        
        #for second inputs
        
        #code for unit of measurement
        if u=='KG_CAP':
            u_exp=' kilograms of retail weight per capita'
        else:
            #one other option
            u_exp=' thousand tonnes of carcass weight'
        
        # convert select values to format for query
        #t.text = "{} consumption of meat in {}".format(s,u_exp)
        s="'{}'".format(s)
        u="'{}'".format(u)
        
        #code for changing era
        if e=="Past":
            sdf=df.query("location=={} and value>0 and time<2020".format(s))
        elif e=="Projected":
            sdf=df.query("location=={} and value>0 and time>2020".format(s))
        else:
            sdf=df.query("location=={} and value>0".format(s))

        sdf_po=sdf.query("subject=='POULTRY'and measure=={}".format(u))
        sdf_be=sdf.query("subject=='BEEF'and measure=={}".format(u))
        sdf_sh=sdf.query("subject=='SHEEP'and measure=={}".format(u))
        sdf_pi=sdf.query("subject=='PIG'and measure=={}".format(u))

        #mx=mdf.time same x values
        po2=sdf_po.value
        be2=sdf_be.value
        sh2=sdf_sh.value
        pi2=sdf_pi.value
        source_avg.data = dict(x=mx, y1=po, y2=be, y3=sh, y4=pi, y5=po2, y6=be2, y7=sh2, y8=pi2) #updated so that changing data source
    
    #updated on chage
    stat_select.on_change('value', update_data_source)
    point_select.on_change('value', update_data_source)
    era_select.on_change('value', update_data_source)
    stat_select2.on_change('value', update_data_source)
    point_select2.on_change('value', update_data_source)
    #era_select2.on_change('value', update_data_source)

    # Set up layouts/adding to doc
    inputs = widgetbox(stat_select, point_select, era_select)
    inputs2= widgetbox(stat_select2,point_select2)
    t.text = "USA consumption of meat in kilograms of retail weight per capita"
    layout = column(t,row(inputs,plot,plot2,plot3,plot4, width=700),row(inputs2,plot5,plot6,plot7,plot8))
    doc.add_root(layout)
    

show(modify_doc)
```
